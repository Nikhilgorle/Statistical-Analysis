## Left Tailed Test and Right Tailed Test

```python
import pandas as pd
from scipy import stats
import statistics
import random

# Load the data from your CSV file
# csv_file_path = 'C:\Users\HP\Dropbox\PC\Desktop\Unit-2\Climate_change.csv'  # Replace with the actual path to your CSV file
df = pd.read_csv('./Climate_change.csv')

# Extract the 'temperature' column as your population data
population_data = df['Temperature']

# Define the sample size you want to use
sample_size = 500  # Adjust the sample size as needed

# Randomly select a sample from the population
sample_data = random.sample(list(population_data), sample_size)

# Calculate Mean (Î¼), Population Proportion (p), and Variance (Ïƒ^2) for the sample
print(f"population mean is: {statistics.mean(population_data)}")
mean = statistics.mean(sample_data)
population_proportion = (sum(1 for x in sample_data if x > 16) / sample_size)  # Adjust threshold_value as needed
variance = statistics.variance(sample_data)

# Left-tailed Hypothesis Test with Î±=0.05 for the sample
null_hypothesis_mean = 14.9360  # Define your null hypothesis mean
alternative_hypothesis_mean = 14.90  # Define your alternative hypothesis mean

t_statistic, p_value = stats.ttest_1samp(sample_data, null_hypothesis_mean)
print(f"p_value : {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (left-tailed test)")
else:
    print("Fail to reject the null hypothesis(left-tailed test)")

# Right-tailed Hypothesis Test with Î±=0.05 for the sample
null_hypothesis_mean = 14.9360  # Define your null hypothesis mean
alternative_hypothesis_mean = 14.90  # Define your alternative hypothesis mean

t_statistic, p_value = stats.ttest_1samp(sample_data, null_hypothesis_mean)
print(f"p_value : {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (right-tailed test)")
else:
    print("Fail to reject the null hypothesis (right-tailed test)")


# Confidence Interval for the sample
confidence_level = 0.95  # Adjust the confidence level as needed
standard_error = stats.sem(sample_data)
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, sample_size - 1) * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print("Confidence Interval:", confidence_interval)

# Z-Test for a Mean Î¼ for the sample
null_hypothesis_mean = 15.0  # Define your null hypothesis mean
alternative_hypothesis_mean = 15.20  # Define your alternative hypothesis mean

sigma = statistics.stdev(sample_data)  # Sample standard deviation
z_score = (mean - null_hypothesis_mean) / (sigma / (sample_size ** 0.5))

p_value = stats.norm.cdf(z_score)

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
