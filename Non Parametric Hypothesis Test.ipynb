## **Runs Test:**

A researcher collected climate change data from various locations across different countries for a specific year,

recording measurements of temperaturein chronological order.

A sample of 20 consecutive data points was selected for analysis.

H0:the data was picked randomly

H1: the data was not picked randomly

alpha=5%

```python
import pandas as pd
import scipy.stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("./Climate_change.csv")
```

```python
temp=[]
import csv

with open("./Climate_change.csv", newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        temperature = row['Temperature']
        temp.append(float(temperature))
print(temp)
```

### using convenience sampling - picking first 20 we see


```python
sample=temp[:20]
print("sample is :",sample)
```

```python
import copy
original_sample=copy.deepcopy(sample)
sample.sort()

median_index=len(sample)//2
median=sample[median_index]

print("sorted sample is :",sample)
print("median index is :",median_index," and median value is :",median)
```

```python
runs=[]

for i in original_sample:
  if i>median:
    runs.append('A')
  elif i<median:
    runs.append('B')
  else:
    runs.append('0')

print("the runs sequence is :",runs)
```

```python
beg=runs[0]
num_runs=0
num_zero=0
num_A=0
num_B=0
for i in runs:
  if i=='A':
    num_A+=1
  elif i=='B':
    num_B+=1

  if i == beg and i!='0':
    continue
  elif i==beg and i=='0':
    beg=i
    num_zero+=1
  elif i!=beg and i!='0':
    num_runs+=1
    beg=i
  elif i!=beg and i=='0':
    num_runs+=1
    num_zero+=1
    beg=i
  else:
    print("exception case")

print("number of runs is :",num_runs)
print("number of zeros in runs sequence is :",num_zero)
print("number of As are :",num_A)
print("number of Bs are :",num_B)
```

```python
R=num_runs
n1=num_A
n2=num_B
```

### Since n1=9 and n2=10 , using the table for alpha=5% we get r1 and r2 as : 5, 16

```python
r1=5
r2=16

if r1<=R<=r2:
  print("Accept H0, the data was selected randomly")
else:
  print("Reject H0, Accept H1 , the data was not picked randomly")
```

# **Paired Sign Test:**

H0: no difference in the humidity level in cities in India before 2010 and after

H1: there is a difference in the humidity level in cities in India before 2010 and after
alpha=5%

calculating 95% confidence interval



Humidity values before 2010:

```python
import csv

humidity_b2010= []

with open('./Climate_change.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    count = 0
    for row in reader:
        if row['Country'] == 'India':
            count += 1
            if count <= 4719 and len(humidity_b2010) < 15:
                humidity = row['Humidity']
                humidity_b2010.append(float(humidity))
print(humidity_b2010)
```

Humidity values after 2010:

```python
import csv

humidity_a2010 = []

with open('./Climate_change.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    count = 0
    for row in reader:
        if row['Country'] == 'India':
            count += 1
            if count > 15 and len(humidity_a2010) < 15:
                humidity = row['Humidity']
                humidity_a2010.append(float(humidity))
print(humidity_a2010)
```

Calculating Sign from difference before and after:


```python
humiditybefore = [int(value) for value in humidity_b2010]
humidityafter =[int(value) for value in humidity_a2010]

```

```python
pos=0
neg=0
zero=0

for i in range(len((humiditybefore))):
  if humiditybefore[i]-humidityafter[i]>0:
    pos+=1
  elif humiditybefore[i]-humidityafter[i]<0:
    neg+=1
  else:
    zero+=1

sign_calc=min(pos,neg)
print("sign calculated value is:",sign_calc)

print("number of samples = ",len(humidity_a2010)-zero)
```

### Since we are doing 2 tail test at alpha=5% , the sign tabulated value for a sample of size 15 is : 3

```python
sign_tab=3

if sign_calc>sign_tab:
  print("Accept HO")
else:
  print("Reject HO, Accept H1")
```

# **Wilcoxon rank sum test**

H0: no difference in the wind speed of cities in India and China

H1: there is difference in the wind speed of cities in India and China


testing this claim with a sample of size 20

alpha=1%



Wind Speed in Cities in China taking sample 20

```python
import csv

windspeedC = []

with open('./Climate_change.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['Country'] == 'China':
            if len(windspeedC) < 20:
                windspeed = row['Wind Speed']
                windspeedC.append(float(windspeed))
windspeedChina = [int(value) for value in windspeedC]

print(windspeedChina)
```

Wind Speed in Cities in India taking sample 15


```python
import csv

windspeedI = []

with open('./Climate_change.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['Country'] == 'India':
            if len(windspeedI) < 15:
                windspeed = row['Wind Speed']
                windspeedI.append(float(windspeed))
windspeedIndia = [int(value) for value in windspeedI]

print(windspeedIndia)
```

### finding the ranks:

```python
vals=[]
for i in range(len(windspeedChina)):
  vals.append(windspeedChina[i])
for i in range(len(windspeedIndia)):
  vals.append(windspeedIndia[i])

vals.sort(reverse=True)

ranks={}
rank=1
for i in vals:
  ranks[i]=rank
  rank+=1

print(ranks)
```

```python
avg_ranks={}

for i in ranks:
  all_ranks=[]
  for j in ranks:
    if j==i and j not in avg_ranks:
      all_ranks.append(ranks[j])

  summ=sum(all_ranks)
  avg_rank=summ/len(all_ranks)
  avg_ranks[i]=avg_rank

print(avg_ranks)
```

Selecting the minimum number of values

```python
print("number of Chinese Cities:",len(windspeedChina))
print("number of Indian cities:",len(windspeedIndia))
n=min(len(windspeedChina),len(windspeedIndia))
print(n)
```

### here the minimum number of values is in India cities so using those:

```python
n1=len(windspeedChina)
n2=len(windspeedIndia)
R=0
for i in range(len(windspeedIndia)):
  R+=avg_ranks[windspeedIndia[i]]

print("R is : ",R)
Mu_R=(n*(n1+n2+1))/2

print("Mu_R is :",Mu_R)
import math

Sigma_R=math.sqrt(((n1*n2)*(n1+n2+1))/12)

print("Sigma R is :" , Sigma_R)
Z_cal=(R-Mu_R)/(Sigma_R)
print("Z_CAL is",Z_cal)
```

### at alpha=1% and 2 tail test , Z_tabulated value is : 2.576

```python
Z_tab=2.576

if Z_cal<=Z_tab:
  print("Accept H0")
else:
  print("Reject H0, Accept H1")
```

alpha=5%

calculating 95% confidence interval

### calculating sign

### Calculating difference and rank

```python
diff=[]

for i in range(len(CO2after)):
  diff.append(abs(CO2after[i]-CO2before[i]))

diff.sort()

avg_ranks={}

for i in range(len(diff)):
  all_ranks=[]
  for j in range(i,len(diff)):
    if diff[j]==diff[i] and diff[j] not in avg_ranks:
      all_ranks.append(j+1)

  if len(all_ranks)>0:
    summ=sum(all_ranks)
    avg_rank=summ/len(all_ranks)
    avg_ranks[diff[i]]=avg_rank

print("ranks are : ", avg_ranks)
diff2=[]
for i in diff:
  diff2.append([i,avg_ranks[i]])

print(diff2)
ws=0
for i in range(len(signs)):
  if signs[i]==sign_calc:
    ws+=diff2[i][1]

print("ws is :",ws)
n=num_samples

mu_ws=(n*(n+1))/4
print("mu_ws is :",mu_ws)
import math

sigma_ws=math.sqrt((n*(n-1)*((2*n)+1))/24)

print("sigma_ws is :",sigma_ws)
z_calc=abs(ws-mu_ws)/(sigma_ws)

print("z calculated is : ",z_calc)
```

### at alpha = 5% and 2 tail test the Z_tabulated value is : 1.960

```python
z_tab=1.960

if z_calc<=z_tab:
  print("Accept H0")
else:
  print("Reject H0, Accept H1")
```
